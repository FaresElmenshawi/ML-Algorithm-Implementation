# Create Data

```python
np.random.seed(1) #every time you call the numpy's other random function, the result will be the same

#Now let's create small arrays to demonstrate the differance
x_small = np.random.rand(10000)  
y_small = np.random.rand(10000)
#Now let's create large arrays to demonstrate the differance
x_large = np.random.rand(100000000)  
y_large = np.random.rand(100000000)
```




# Define Dot Function

```python
def my_dot(x, y): 
    total = 0
    for i in range(x.shape[0]):
        total = total + x[i] * y[i]
    return total
```

# Compare  Implementations

## Compare small data difference

```python
start_time = time.time() #capture start time
np.dot(x_small, y_small)
end_time = time.time() #capture end time
print(f"Vectorized version duration: {1000*(end_time-start_time):.2f} ms ")

start_time = time.time() #capture start time
my_dot(x_small, y_small)
end_time = time.time() #capture end time
print(f"Non-Vectorized version duration: {1000*(end_time-start_time):.2f} ms ")
```

## Compare large data difference

```python
start_time = time.time() #capture start time
np.dot(x_large, y_large)
end_time = time.time() #capture end time
print(f"Vectorized version duration: {1000*(end_time-start_time):.2f} ms ")

start_time = time.time() #capture start time
my_dot(x_large, y_large)
end_time = time.time() #capture end time
print(f"Non-Vectorized version duration: {1000*(end_time-start_time):.2f} ms ")
```
